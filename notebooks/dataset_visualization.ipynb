```python
def read_file(path: str):
    """Функция для считывания файла .txt"""
    if not os.path.exists(path):
        return []
    with open(path, "r", encoding="utf-8") as f:
        return [line.strip() for line in f if line.strip()]
```

```python
def read_data(path: str) -> pd.DataFrame:
    
    # open patients folders
    parients_listdirs = os.listdir(path)
    assert len(parients_listdirs) == 42, f"patients dirs: {len(parients_listdirs)}"
    
    # Для каждого пациента
    # lesionVideos.txt - видео с поражениями
    # nonlesionVideos.txt - видео без поражений

    for item_patient in parients_listdirs:

        path2item_patient = f"{path}/{item_patient}"
        
        # === Видео где есть поражения ===
        video_lesionVideos = read_file(f"{path2item_patient}/lesionVideos.txt")
        # === Видео без поражений ===
        video_nonlesionVideos = read_file(f"{path2item_patient}/nonlesionVideos.txt")

        
```
